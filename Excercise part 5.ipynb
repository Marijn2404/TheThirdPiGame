Read Excel data into Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_excel("BakeryData_Vilnius.xlsx")
stores = ["main street A","main street B","station A","station B"]
df["weekday"] = df['date'].dt.strftime('%A')
df["date"] = pd.to_datetime(df["date"]).dt.date
df
```

Summary statistics for the dataset

```python
df.describe()
```

```python
for store in stores:
    fig = px.histogram(df, x=store, color="weekday", title="Histogram for {0}".format(store))
    fig.show()
```

```python
for store in stores:
    fig = px.box(df, x=store, color="weekday", title=store)
    fig.show()
```

```python
for store in stores:
    fig = px.line(df, x="date", y=store, color="weekday", title=store)
    fig.show()
```
