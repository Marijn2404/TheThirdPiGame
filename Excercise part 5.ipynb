Read Excel data into Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("BakeryData_Vilnius.xlsx")
df
```

Summary statistics for the dataset

```python
df.describe()
```

```python
def histogram(axs, number, name):
    axs[number].hist(df[name])
    axs[number].set_title("Histrogram for {0}".format(name))

fig, axs = plt.subplots(1,4, figsize=(15,10))

names = ["main street A","main street B","station A","station B"]

for i in enumerate(names):
    histogram(axs, i[0],i[1])

plt.tight_layout()
plt.show()
```
